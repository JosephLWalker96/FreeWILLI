```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal_length = 10000

click_dur = 100

mu, sigma = 0, 10 # mean and standard deviation
s = np.abs(np.random.normal(mu, sigma, signal_length))

for i in range(10):
    new_s = s.copy()
    rand_num = np.random.rand(1)
    print(rand_num)
    if rand_num < .999:
        start_pos = int(np.random.rand(1)*signal_length)
        new_s[start_pos:start_pos + click_dur] = (new_s[start_pos:start_pos + click_dur] * 1.7) * (np.hamming(click_dur) + 1)
        new_s = np.array(new_s, dtype=np.uint16)
    plt.plot(np.arange(0,len(new_s),1),new_s)
    plt.show()
```
