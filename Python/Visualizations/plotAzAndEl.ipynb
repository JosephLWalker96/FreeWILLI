```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the column names
column_names = ["usec_since_Unix_Start", "Energy", "El.", "Az.", "TDOA12", "TDOA13", "TDOA14", 
                "TDOA23", "TDOA24", "TDOA34", "Xcorr12", "Xcorr13", "Xcorr14", 
                "Xcorr23", "Xcorr24", "Xcorr34"]

# Load the data with the specified column names
path_doa = "../../C/deployment_files/2024-8-29-18-37-36-480140_detection"
df = pd.read_csv(path_doa, delim_whitespace=True, names=column_names, skiprows=1)

# Display the first few rows to verify
print(df.head())

# Plot Azimuth vs. Time with transparency
plt.figure(figsize=(10, 5))

azData = df["Az."]

# Set transparency using alpha
plt.scatter(df["usec_since_Unix_Start"], azData, marker='o', s=3, color='b', alpha=0.2)

plt.title("Azimuth vs. Time")
plt.xlabel("Time (usec since Unix Start)")
plt.ylabel("Az2")
plt.grid(True)
plt.show()

# Plot Elevation vs. Time with transparency
plt.figure(figsize=(10, 5))
plt.scatter(df["usec_since_Unix_Start"], df["El."], marker='o', s=3, color='b', alpha=0.2)

plt.title("Elevation vs. Time")
plt.xlabel("Time (usec since Unix Start)")
plt.ylabel("El2")
plt.grid(True)
plt.show()


```
