```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage.transform import hough_line, hough_line_peaks

# Load the data as per your existing code
column_names = ["usec_since_Unix_Start", "Energy", "El.", "Az.", "TDOA12", "TDOA13", "TDOA14", 
                "TDOA23", "TDOA24", "TDOA34", "Xcorr12", "Xcorr13", "Xcorr14", 
                "Xcorr23", "Xcorr24", "Xcorr34"]

path_doa = "../../C/deployment_files/2024-8-29-18-37-36-480140_detection"
df = pd.read_csv(path_doa, delim_whitespace=True, names=column_names, skiprows=1)

# Function to run the Hough Transform and visualize the results
def run_hough_transform(df):
    # Extract relevant features for Hough Transform
    time = df["usec_since_Unix_Start"].values
    azimuth = df["Az."].values
    
    # Combine features into 2D points (time, azimuth)
    points = np.vstack((time, azimuth)).T
    
    # Normalize the data for the Hough Transform
    points -= points.mean(axis=0)
    
    # Adjust the range and resolution for the Hough Transform
    tested_angles = np.linspace(-np.pi / 2, np.pi / 2, 3600)  # Increase the number of angle samples
    hspace, angles, dists = hough_line(points, theta=tested_angles)
    
    # Identify the peaks in the Hough space
    accum, angles, dists = hough_line_peaks(hspace, angles, dists, threshold=0.03 * np.max(hspace),
                                            min_distance=15, num_peaks=20)

    print("accum ", accum)
    print("angles ", angles)
    print("dists ", dists)
    
    # Initialize labels for visualization
    labels = np.full(len(df), -1)
    label_counter = 0
    
    # Iterate over the detected lines (peaks in Hough space)
    for angle, dist in zip(angles, dists):
        # Calculate the cosine and sine of the angle
        cos_theta = np.cos(angle)
        sin_theta = np.sin(angle)
        
        # The equation of the line in the original (time, azimuth) space is:
        # dist = time * cos(theta) + azimuth * sin(theta)
        # Rearrange to find points close to this line:
        line_indices = np.abs(points[:, 0] * cos_theta + points[:, 1] * sin_theta - dist) < 1.0
        
        line_points = points[line_indices]
        print("line_points ", line_points)
        
        # Assign a unique label to the points belonging to this detected line
        labels[line_indices] = label_counter
        
        label_counter += 1
    
    # Plot the results
    plt.figure(figsize=(10, 6))
    plt.scatter(df["usec_since_Unix_Start"], df["Az."], marker='o', c=labels, cmap='viridis')
    plt.title("Azimuth vs. Time with Detected Tracks")
    plt.xlabel("Time (usec since Unix Start)")
    plt.ylabel("Azimuth")
    plt.grid(True)
    plt.colorbar(label='Track ID')
    plt.show()

# Periodically run the Hough Transform process
time_intervals = np.linspace(df["usec_since_Unix_Start"].min(), df["usec_since_Unix_Start"].max(), num=20)
for ind in range(len(time_intervals) - 1):
    # Select the data within the current interval
    df_current = df[(df["usec_since_Unix_Start"] > time_intervals[ind]) & (df["usec_since_Unix_Start"] <= time_intervals[ind+1])]
    
    # If there are no data points in this interval, skip to the next iteration
    if df_current.empty:
        continue
    
    # Run the Hough Transform and visualize the results
    run_hough_transform(df_current)
```
