```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
import struct
```

```python
# Replace 'your_file_path.mat' with the path to your .mat file
file_path = '../Data/joesdata.mat'

# Load the .mat file
DATA = loadmat(file_path)['DATA']
print(DATA.shape)
DATA = DATA.T
```

```python
num_rows = DATA.shape[0]

fig, axs = plt.subplots(num_rows, 1)  # Adjust figsize as needed

# Plotting each row as a subplot
for i in range(num_rows):
    axs[i].plot(DATA[i])
    #axs[i].set_title(f'Row {i+1}')

# Adjust layout
#plt.tight_layout()

# Show the plot
plt.show()
```

```python
def detect_pulses(audio_signal, noise_threshold=90, pulse_threshold=150, pulse_duration=100):
    print('HERE ',len(audio_signal))
    above_noise = audio_signal > noise_threshold
    pulse_filter = np.ones(pulse_duration, dtype=bool)
    above_pulse_threshold = np.convolve(audio_signal > pulse_threshold, pulse_filter, mode='valid') == pulse_duration

    pulse_starts = np.where(above_noise & above_pulse_threshold)[0]
    pulse_ends = pulse_starts + pulse_duration

    pulse_intervals = []
    for start, end in zip(pulse_starts, pulse_ends):
        pulse_intervals.append((start, end))

    return pulse_intervals
```

```python
ch1_data = DATA[0]
length = len(ch1_data)
samples_per_segment = 50000

num_segments = length // samples_per_segment

for i in range(num_segments):
    samps = ch1_data[i*samples_per_segment:(i+1)*samples_per_segment]
    samps = np.power(samps, 2, dtype=np.int64)
    #print(np.amax(samps),np.amin(samps))
    samps = np.sqrt(samps)
    #print(np.amax(samps),np.amin(samps))
    plt.plot(samps)
    plt.show()
    
    pulses = detect_pulses(samps)
    print(pulses)
    
```
