```python
from scipy.io import loadmat
import matplotlib.pyplot as plt
import numpy as np
import struct
```

```python
# Replace 'your_file_path.mat' with the path to your .mat file
file_path = '../Data/joesdata.mat'

# Load the .mat file
DATA = loadmat(file_path)['DATA']
print(DATA.shape)
DATA = DATA.T
```

```python
num_rows = DATA.shape[0]

fig, axs = plt.subplots(num_rows, 1)  # Adjust figsize as needed

# Plotting each row as a subplot
for i in range(num_rows):
    axs[i].plot(DATA[i])
    #axs[i].set_title(f'Row {i+1}')

# Adjust layout
#plt.tight_layout()

# Show the plot
plt.show()
```

## Encode and Decode the data

```python
sub_data = DATA[:,:15]
for row in sub_data:
    print(row)
```

```python
# Interleave the values of the rows evenly to create a 1x40 vector
sub_data_reshaped = sub_data.reshape(-1, order='F')
print(sub_data_reshaped.shape)
print(sub_data_reshaped)
```

```python
scaled = sub_data_reshaped + 2**15
print(scaled)
print(scaled.shape)
```

```python
converted = np.array(scaled,dtype=np.uint16)
print(converted)
```

```python
data_bytes = converted.tobytes() # big_endian
```

```python

decoded_array = np.array(struct.unpack('<' + 'H' * (len(data_bytes) // 2), data_bytes))
print(decoded_array)

## or alternatively:
decoded_array = np.frombuffer(data_bytes, dtype=np.uint16)
print(decoded_array)
```

```python
rescaled = np.array(decoded_array - 2**15).astype(np.int16)
print(rescaled)
```
